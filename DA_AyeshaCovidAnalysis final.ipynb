```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('OxCGRT_compact_national_v1.csv')
```

```python
data.head(5)
```

```python
data = data.sort_values(by=['ConfirmedCases'],ascending = False)
data
```

```python
testdata = data[data['CountryName'] == 'India']
testdata
```

```python
tempdata = data
tempdata = tempdata[tempdata['Date'] == 20221231]
finaldata = tempdata[['CountryName','ConfirmedDeaths']]
```

```python
finaldata = finaldata.sort_values(by=['ConfirmedDeaths'],ascending = False)
finaldata['CountryName'] = finaldata['CountryName'].replace({'United States': 'United States of America'})
finaldata.head(10)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
world = gpd.read_file('110m_cultural/ne_110m_admin_0_countries.shp')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1, color='k')
plt.show()
```

```python
# Load the world GeoDataFrame
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world GeoDataFrame with your data using the 'CountryName' column
merged_data = world.set_index('name').join(finaldata.set_index('CountryName'))

# Plot the data on a world map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1, color='k')
merged_data.plot(column='ConfirmedDeaths', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

plt.title('Confirmed Deaths by Country')
plt.show()
```

```python
int_popu = pd.read_csv('2021_population.csv')
```

```python
int_popu
```

```python
popu = int_popu[['country','2021_last_updated']]
popu['country'] = popu['country'].replace({'United States': 'United States of America'})
```

```python
popu.head(5)
```

```python
data = pd.merge(finaldata, popu, how='inner', left_on='CountryName', right_on='country')
data = data.drop(columns=['country'])
```

```python
data = data.dropna(subset=['ConfirmedDeaths'])
data
```

```python
data_type = data['2021_last_updated'].dtypes
data_type
data['2021_last_updated'] = data['2021_last_updated'].str.replace('[^\d]', '', regex=True).astype(int)
```

```python
data['2021_last_updated'] = pd.to_numeric(data['2021_last_updated'], errors='coerce', downcast='integer')
data['Population(Millions)'] = data['2021_last_updated'] / 10**6
data['DeathsByPopulation'] = data['ConfirmedDeaths'] / data['Population(Millions)']
```

```python
data
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1, color='k')
plt.show()
```

```python
# Load the world GeoDataFrame
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world GeoDataFrame with your data using the 'CountryName' column
merged_data = world.set_index('name').join(data.set_index('CountryName'))

# Plot the data on a world map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1, color='k')
merged_data.plot(column='DeathsByPopulation', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

plt.title('Population Averaged Deaths By Country')
plt.show()
```
